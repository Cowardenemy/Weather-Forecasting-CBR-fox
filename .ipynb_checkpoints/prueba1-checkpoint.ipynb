```python
import pandas as pd
import numpy
import numpy as np
import math
from numpy import hstack
from numpy import array
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense, LSTM, Dropout
import matplotlib.pyplot as plt
from sklearn.model_selection import TimeSeriesSplit
```

```python
df_temp = pd.read_csv("DatosMeteorologicos.csv", parse_dates= True, index_col= 1)
df_temp.head()
```

```python
#leer de nuevo el csv para que las grficas aparezcan por aos
df_temp_1 = pd.read_csv("DatosMeteorologicos2.csv", parse_dates=True, index_col=0)
muestra = df_temp_1['2000':]
plt.figure(figsize=(20,8))
muestra['PRES_MAX'].plot(legend = True)
muestra['PRES_MED'].plot(legend = True)
muestra['PRES_MIN'].plot(legend = True)
plt.title('Grfica de la presin')
plt.ylabel('Presin')
plt.xlabel('Aos')
plt.legend(['Presin mxima','Presin media','Presin mnima'])
#plt.savefig("Temperatura.png")
plt.show()
```

```python
def split_sequences(sequences, n_steps):
	X, y = list(), list()
	for i in range(len(sequences)):
		# find the end of this pattern
		end_ix = i + n_steps
		# check if we are beyond the dataset
		if end_ix > len(sequences):
			break
		# gather input and output parts of the pattern
		seq_x, seq_y = sequences[i:end_ix, :-1], sequences[end_ix-1, -1]
		X.append(seq_x)
		y.append(seq_y)
	return array(X), array(y)

values_train = df_temp.filter(['TEMP_MAX','TEMP_MED','TEMP_MIN'])
data_values = values_train.values

values_predict = df_temp.filter(['TEMP_MAX'])
data_predict = values_predict.values

step_days = 14

dataset = hstack((data_values, data_predict))
input1, target = split_sequences(dataset, step_days)
```

```python
input_train, input_test, target_train, target_test = train_test_split(input1, target, test_size = 0.30, random_state=4,shuffle=True)
```

```python
model = Sequential()
model.add(LSTM(100, activation = 'relu', return_sequences = True, input_shape = (step_days ,input_train.shape[2])))
model.add(LSTM(32, activation = 'relu', return_sequences=False))
model.add(Dropout(0.2))
model.add(Dense(1))#, activation='linear'
model.compile(optimizer = 'adam', loss = 'mse')
history = model.fit(input_train, target_train, validation_data = (input_test, target_test),batch_size = 16, epochs = 100)
```

```python
plt.figure(figsize=(40,8))
for i in range(0,100):
  history.history['loss'][i]= math.sqrt( history.history['loss'][i])
  history.history['val_loss'][i]=math.sqrt( history.history['val_loss'][i])
plt.plot( history.history['loss'], '-o')
plt.plot(history.history['val_loss'], '-o')
plt.title('RMSE')
plt.ylabel('RMSE value')
plt.xlabel('No. epoch')
plt.legend(loc="upper left")
print(history.history['loss'][99])
print(history.history['val_loss'][99])
plt.show()
```

```python

muestrav = df_temp.filter(['TEMP_MAX'])
muestra = muestrav.values
print(muestra)
```

```python
arreglo=np.array(muestra)
win_size=14
n_slic=len(target)-2*win_size
#slic=np.zeros( (n_slic,win_size) )
slic=[]

def win(i, win_size):
    #Esto es 300% optimizable
    corte=slice(i,i+win_size)
    datos=arreglo[corte]
    datos=np.array(datos)
    return datos

for i in range(0,len(target)-win_size):
    slic.append(win(i,win_size))

slic=np.asarray(slic)
print(slic)
```

```python
def comp(target,slic):
    target2 = [] 
    slic2 = []
    for i in range(len(target)):
        target2.append(target[i][0])
        slic2.append(slic[i][0])        
    my_rho = np.corrcoef(target2, slic2)
    dist = np.linalg.norm(target-slic)
    return my_rho[0][1], dist

def comp_p2(pearson,euclides):
    dato=((1-euclides+pearson)/2)*100
    return dato

pearson=np.zeros(len(slic))
euclides=np.zeros(len(slic))
final_pearson_euclides=np.zeros(len(slic))

for i in range(len(slic)):
    pearson[i],euclides[i]=comp(slic[len(slic)-1],slic[i])

maximo=max(euclides)
euclides=euclides/maximo

for i in range(len(slic)):
    final_pearson_euclides[i]=comp_p2(pearson[i],euclides[i])
#print(pearson_euclides)
print(final_pearson_euclides)
#print(final_pearson_euclides.size)
```
